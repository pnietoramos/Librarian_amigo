### Extracting fragments for all authors and merging them together in a single pikle file

#### Loading file names from ataun

```python
result = subprocess.check_output("ls all_authors", shell=True)
file_names = result.decode(encoding='UTF-8').split("\n")[:-1]
file_names[0:5]
```

#### Removing headers and hyphenation. Slicing each book in slices of ten sentences an making a dictionary with all books. (Ataun)

```python
all_books ={}
for name in file_names:
    with open("all_authors/" + name,"r") as file:
        string_file = "".join(file.read().split("www.luarna.com")[1:])
        string_file = string_file.lower()
        remove = ['\x0c', '-\n\n', '-\n', '!', '¡', '¿', '\?', '\(', '\)']
        for item in remove:
            string_file = re.sub(item, '', string_file)
        to_space = [' vd. ', ' sr. ', ' d. ', '\..']
        for item in to_space:
            string_file = re.sub(item, ' ', string_file)
        string_file = re.sub('-', '.', string_file)
        string_file = " ".join(string_file.split("\n"))        
        name = name[:-4]
        all_books[name]=[]
        fragment = string_file.split(".")
        for i in range(len(fragment)//5):
            all_books[name].append(" ".join(fragment[(i)*5:(i+1)*5]))
            
```
